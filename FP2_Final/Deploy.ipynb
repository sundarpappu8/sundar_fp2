# Applying Auto TS for Transportation Sector

```python
df1['ds'] = pd.to_datetime(df1.ds)
```

```python
df1['ds'] = df1['ds'].dt.strftime('%d-%b-%Y') # change time formatting
print(df1['ds'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#split the data into train and test set
train_df1, test_df1 = train_test_split(df1[:], test_size=0.2, random_state=0, shuffle=False)
```

```python
train_df1
```

```python
test_df1
```

```python
df1_model = auto_timeseries(score_type='rmse', 
                            time_interval= None, 
                            non_seasonal_pdq=None, 
                            seasonality=True, 
                            seasonal_period=12, 
                            model_type=['best', 'prophet', 'stats', 'ARIMA', 'SARIMAX', 'VAR', 'ML'], 
                            forecast_period=10,
                            verbose=2)
```

```python
df1_model.fit(traindata=train_df1, 
              ts_column='ds', 
              target='y')
              #cv=5, 
              #sep="," )
```

```python
df1_model.get_leaderboard()
```

```python
df1_model.plot_cv_scores()
```

```python
df1_model.get_cv_scores()
```

```python
results_dict = df1_model.get_ml_dict()
```

```python
results_dict['auto_SARIMAX'] 
```

```python
prediction_df1 = df1_model.predict(testdata=12,model='auto_SARIMAX',simple=False)
```

```python
prediction_df1 
```

```python
df1_model.get_best_model()
```

```python
SARIMAX_model = df1_model.get_model('auto_SARIMAX')
```
