#Decomposition of Consumption timeseries for each sector 

```python
data_decompose=df_monthly[:]
data_decompose.head(5)
```

```python
data_decompose.set_index('Month',inplace=True)
data_decompose.index=pd.to_datetime(data_decompose.index)
data_decompose
```

```python
data_decompose.info()
```

```python
print(pd.infer_freq(data_decompose))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result_Commercial_Sector = seasonal_decompose(data_decompose['Commercial Sector'], model='multiplicative', period= 12)
fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize=(15,12))
result_Commercial_Sector.trend.plot(ax=ax1)
result_Commercial_Sector.seasonal.plot(ax=ax2)
result_Commercial_Sector.resid.plot(ax=ax3)
```

```python
result_End_Use_Sectors = seasonal_decompose(data_decompose['End-Use-Sectors'], model='multiplicative', period= 12)
fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize=(15,12))
result_End_Use_Sectors.trend.plot(ax=ax1)
result_End_Use_Sectors.seasonal.plot(ax=ax2)
result_End_Use_Sectors.resid.plot(ax=ax3)
```

```python
result_Industrial_Sector  = seasonal_decompose(data_decompose['Industrial Sector'], model='multiplicative', period= 12)
fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize=(15,12))
result_Industrial_Sector.trend.plot(ax=ax1)
result_Industrial_Sector.seasonal.plot(ax=ax2)
result_Industrial_Sector.resid.plot(ax=ax3)
```

```python
result_Residential_Sector = seasonal_decompose(data_decompose['Residential Sector'], model='multiplicative', period= 12)
fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize=(15,12))
result_Residential_Sector.trend.plot(ax=ax1)
result_Residential_Sector.seasonal.plot(ax=ax2)
result_Residential_Sector.resid.plot(ax=ax3)
```

```python
result_Transportation_Sector = seasonal_decompose(data_decompose['Transportation Sector'], model='multiplicative', period= 12)
fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize=(15,12))
result_Transportation_Sector.trend.plot(ax=ax1)
result_Transportation_Sector.seasonal.plot(ax=ax2)
result_Transportation_Sector.resid.plot(ax=ax3)
```

# Plotting ACF and PACF for each sector

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df_monthly['Commercial Sector'],lags=100)
plot_pacf(df_monthly['Commercial Sector'],lags=50)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df_monthly['End-Use-Sectors'],lags=100)
plot_pacf(df_monthly['End-Use-Sectors'],lags=50)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df_monthly['Industrial Sector'],lags=100)
plot_pacf(df_monthly['Industrial Sector'],lags=50)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df_monthly['Residential Sector'],lags=100)
plot_pacf(df_monthly['Residential Sector'],lags=50)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df_monthly['Transportation Sector'],lags=100)
plot_pacf(df_monthly['Transportation Sector'],lags=50)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df_monthly['Commercial Sector'],lags=100)
plot_pacf(df_monthly['Commercial Sector'],lags=50)
plt.show()
```
